```python
age_counts=df["Age"].value_counts().sort_index()
plt.bar(age_counts.index, age_counts.values, color='lightblue', edgecolor='white')
plt.title('Distribution of Ages', fontsize=16)
plt.xlabel('Ages', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.gcf().patch.set_facecolor('yellow')
plt.gca().set_facecolor('black')
plt.show()
```

```python
gender_counts=df["Gender"].value_counts()
gender_counts
```

```python
columns = df[['Age', 'Gender', 'Marital Status', 'Occupation', 'Monthly Income',
       'Educational Qualifications', 'Family size', 'Output', 'Feedback']]

plt.figure(figsize = (22, 25))
for i, col in enumerate(columns):
    plt.subplot(7, 2, i+1) 
    counts = df[col].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
    plt.title(col)

plt.tight_layout()
plt.show()
```
