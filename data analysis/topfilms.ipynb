```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("imdb_top_2000_movies.csv")
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
df.sample(3)
```

```python
df["Genre"]=df["Genre"].str.replace(",","-")
df.head()
```

```python
df["Votes"]=df["Votes"].str.replace(",",".")
df.sample(1)
```

```python
df.columns=df.columns.str.lower()
df.sample(1)
```

```python
df["genre"]=df["genre"].str.lower()
```

```python
df.sample(1)
```

```python
df["diretor"]=df["director"].str.lower()
df.sample(1)
```

```python
df.rename(columns={
    "movie name":"film ismi",
    "release year":"yayın tarihi",
    "duration":"film süresi",
    "imdb raint":"puan",
    "metascore": "skor",
    "votes":"toplam oy",
    "genre":"tür",
    "director":"yönetmen",
    "cast":"baş rol",
    "gross":"brüt"
}, inplace=True)
df.head(2)
```

```python
df.drop(columns=["diretor"],inplace=True)
df.sample(1)
```

```python
df["yönetmen"]=df["yönetmen"].str.lower()
```

```python
df.head(2)
```

```python
df["baş rol"]=df["baş rol"].str.lower()
df.sample(2)
```

```python
df["brüt"]=df["brüt"].str.replace("$","")
df["brüt"]=df["brüt"].str.replace("M","")
df.sample(2)
```

```python
df.dtypes
```

```python
df[df.yönetmen.str.contains("anne fletcher")]
```

```python
df[df["baş rol"].str.contains("katherine heigl")]
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
ortalama=df["skor"].mean()
df["skor"].fillna(ortalama,inplace=True)
df.sample(1)
```

```python
df["brüt"]=df["brüt"].astype("float64")
```

```python
ort=df["brüt"].mean()
df["brüt"].fillna(ort,inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(data=df, x='skor', y='brüt')
plt.xlabel('yönetmen')
plt.ylabel('Brüt')
plt.title('Skor ve Brüt Arasındaki İlişki')
plt.show()
```

```python
plt.hist(stevenn['imdb rating'], bins=10, color='skyblue', edgecolor='black', alpha=0.7, label='IMDb Puanı')
plt.hist(stevenn['brüt'], bins=10, color='red', edgecolor='black', alpha=0.7, label='Brüt Gelir')
plt.xlabel('Değer')
plt.ylabel('Film Sayısı')
plt.title('Steven Spielberg Filmlerinin IMDb Puanı ve Brüt Gelir Dağılımı')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(stevenn['imdb rating'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('IMDb Puanı')
plt.ylabel('Film Sayısı')
plt.title('Steven Spielberg Filmlerinin IMDb Puanı Dağılımı')

plt.subplot(1, 2, 2)
plt.hist(stevenn['brüt'], bins=10, color='red', edgecolor='black')
plt.xlabel('Brüt Gelir')
plt.ylabel('Film Sayısı')
plt.title('Steven Spielberg Filmlerinin Brüt Gelir Dağılımı')

plt.tight_layout()
plt.show()
```
