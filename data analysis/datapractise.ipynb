```python
df_ulke=df.groupby("Ülke")
```

```python
df_ulke_sayi=pd.DataFrame(df_ulke.size().sort_values(ascending=False),
                         columns=["Sayi"])
# df_ulke.size(): Bu, DataFrame'deki her bir benzersiz ülkenin ne kadar sıklıkla geçtiğini hesaplar.
# size() fonksiyonu, her bir benzersiz değerin kaç kez geçtiğini sayar.
# .sort_values(ascending=False): Bu, sıralama işlemidir.
# sort_values() fonksiyonu, belirtilen sütuna göre DataFrame'i sıralar.
# ascending=False parametresi, sıralamanın azalan sırada yapılmasını sağlar,
# yani en yüksek sayıdan en düşük sayıya doğru sıralama yapar.
```

```python
df_ulke_sayi.head()
```

```python
sns.barplot(x = df_ulke_sayi["Sayi"][:10], y = df_ulke_sayi.index[:10])
```
