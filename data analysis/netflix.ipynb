```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.rename(columns={"show_id":"showid","listed_in":"listedin","date_added":"dateadded","release_year":"relaseyear"}, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df["dateadded"] = df["dateadded"].str.replace(",","-")
```

```python
df.head()
```

```python
df["listedin"] = df["listedin"].str.replace(",","-")
```

```python
df.tail()
```

```python
df["description"] = df["description"].str.replace(",","-")
df.tail()
```

```python
df.head()
```

```python
df["cast"] = df["cast"].str.replace(",","-")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df["type"].value_counts()
```

```python
nt = df.groupby(["rating"]).size().reset_index(name="counts")
nt
```

```python
sorted_nt = nt.sort_values(by="counts", ascending=False)
sorted_nt
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='counts', data=sorted_nt, palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Counts')
plt.title('Counts of Each Rating Category')
plt.xticks(rotation=45)
plt.show()
```
