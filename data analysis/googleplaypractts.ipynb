```python
rating_median=df["Rating"].median()
rating_median
```

```python
df["Rating"].fillna(rating_median, inplace=True)
```

```python
df["Rating"] = df["Rating"].astype("int64")
```

```python
df["Size"].unique()
# uyg boyutu
```

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

```python
df["Size"].replace("M","", regex=True, inplace=True)
df["Size"].replace("k","", regex=True, inplace=True)
```

```python
df["Size"].unique()
```

```python
size_median=df[df["Size"]!="Varies with device"]["Size"].astype(float).median()
size_median
```

```python
df["Size"].replace({"Varies with device":size_median}, inplace=True)
```

```python
df.Size=pd.to_numeric(df.Size)
# Bu kod, "Size" sütunundaki değerleri sayısal değerlere dönüştürmek için pandas'ın to_numeric() fonksiyonunu kullanır.
# İşte bu işlemi adım adım açıklayalım:
# df.Size: Bu, DataFrame içindeki "Size" sütununu seçer. "Size" sütunu üzerinde işlem yapılacak olan sütundur.
# pd.to_numeric(): Bu, pandas kütüphanesinin to_numeric() fonksiyonudur.
# Bu fonksiyon, bir dizi veya DataFrame sütununu sayısal değerlere dönüştürmek için kullanılır.
# pd.to_numeric(df.Size): Bu ifade, "Size" sütununu sayısal değerlere dönüştürür.
# Eğer "Size" sütununda sayısal olmayan değerler varsa, bu değerler NaN (Not a Number) olarak işaretlenir.
```

```python
df.Installs=df.Installs.apply(lambda x:x.replace("+",""))
df.Installs=df.Installs.apply(lambda x:x.replace(",",""))
df.Installs=df.Installs.apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

```python
df["Price"].unique()
```

```python
df.Price=df.Price.apply(lambda x:x.replace("$",""))
df.Price=df.Price.apply(lambda x:float(x))
```
